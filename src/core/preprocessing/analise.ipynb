```python
def normalize(info):
    return info.replace(".", '').replace(',','.')[3:]
```

```python
df3['comissão'] = df3['comissão'].apply(normalize).astype(float)
df3['cashback'] = df3['cashback'].apply(normalize).astype(float)
df3['vendas totais'] = df3['vendas totais'].apply(normalize).astype(float)
```

```python
df2['comissão'] = df2['comissão'].apply(normalize).astype(float)
df2['cashback'] = df2['cashback'].apply(normalize).astype(float)
df2['vendas totais'] = df2['vendas totais'].apply(normalize).astype(float)
```

```python
df2.groupby('Grupos de usuários')[['comissão', 'cashback']].sum()
```
